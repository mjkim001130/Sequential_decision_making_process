```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
L = np.array([[1, -1, 0],
              [-1 ,2 ,-2],
              [2, -1, 1]])

P_theta = np.array([1/5 ,1/5, 3/5])
print(L, end="\n\n")
print(P_theta)
```

### 1. Numerically solve the Game with Nature using Non-Deterministic Nature Model discussed in the class



```python
u_star_nondeterminisitc = np.argmin(np.max(L))
print("u_",u_star_nondeterminisitc, sep="")
```

### 2. Numerically solve the Game with Nature using Probabilistic Nature Model discussed in the class

```python
u_star_probabilistic = np.argmin(np.dot(L, P_theta))
print("u_",u_star_probabilistic, sep="")
```

### Under the Your Agent" vs. Nature game setting, run 1000 trials for each strategy and plot the cumulative cost(y-axis) along the trial(x-axis)

```python
cost1 = []
cost2 = []

for i in range(1000):
    theta = np.random.choice([0, 1, 2], p=P_theta)
    cost1.append(L[u_star_nondeterminisitc, theta])
    cost2.append(L[u_star_probabilistic, theta])

plt.plot(np.cumsum(cost1), label="Nondeterministic")
plt.plot(np.cumsum(cost2), label="Probabilistic")
plt.legend()
plt.show()
```
