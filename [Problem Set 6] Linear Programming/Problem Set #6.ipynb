```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cost_matrix = np.array([[3, 0],
                        [-1, 1]])

cost_matrix
```

### Find $w*$ for P1

```python
def find_optimal_w(cost_matrix):
    w1_values = np.linspace(0, 1, 1000)

    w = np.array([w1_values, 1-w1_values]).T
    s = np.dot(w, cost_matrix)
    s1 = s[:, 0]
    s2 = s[:, 1]

    z1, z2 = 1, 0
    L_z1 = s1*z1 + s2*z2

    z1, z2 = 0, 1
    L_z2 = s1*z1 + s2*z2

    L_worst = np.maximum(L_z1, L_z2)

    min_index = np.argmin(L_worst)

    plt.figure(figsize=(10, 5))
    plt.plot(w1_values, L_z1, label='L(z1) with z_1=1', linestyle='--')
    plt.plot(w1_values, L_z2, label='L(z2) with z_2=1', linestyle='-.')
    plt.plot(w1_values, L_worst, label='L(worst)', color='red')
    plt.xlabel('w1')
    plt.ylabel('L(w,z)')
    plt.legend()
    plt.title('Optimal strategy for P1')
    plt.show()

    return w1_values[min_index], L_worst[min_index]

w_star, L_star_P1 = find_optimal_w(cost_matrix)
w_star = [w_star, 1 - w_star]

print('Optimal w*:', w_star[0], 'with L*: ', L_star_P1)
```

### Find $z*$ for P2

```python
def find_optimal_z():
    z1_values = np.linspace(0, 1, 1000)

    w = np.array([z1_values, 1-z1_values])
    s = np.dot(cost_matrix, w)
    s1 = s[0, :]
    s2 = s[1, :]

    w1, w2 = 1, 0
    L_w1 = w1*s1 + w2*s2

    w1, w2 = 0, 1
    L_w2 = w1*s1 + w2*s2

    L_worst = np.minimum(L_w1, L_w2)

    max_index = np.argmax(L_worst)

    plt.figure(figsize=(10, 5))
    plt.plot(z1_values, L_w1, label='L(z1) with z_1=1', linestyle='--')
    plt.plot(z1_values, L_w2, label='L(z2) with z_2=1', linestyle='-.')
    plt.plot(z1_values, L_worst, label='L(worst)', color='red')
    plt.xlabel('z1')
    plt.ylabel('L(w,z)')
    plt.legend()
    plt.title('Optimal strategy for P2')
    plt.show()

    return z1_values[max_index], L_worst[max_index]

z_star, L_star_P2 = find_optimal_z()
z_star = [z_star, 1 - z_star]

print('Optimal z*:', z_star[0], 'with L*: ', L_star_P2)
```
